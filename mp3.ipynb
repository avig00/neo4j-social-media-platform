```python
# QUERY 4: Find common interests between two specific Users
query4 = '''
    MATCH (user1:User {vunet_id: 'isabelarv'}), (user2:User {vunet_id: 'chenyum'})
    RETURN user1.name, user2.name,
           [interest IN user1.fields_of_interest WHERE interest IN user2.fields_of_interest],
           [generalInterest IN user1.general_interests WHERE generalInterest IN user2.general_interests]
'''

result4 = session.run(query4)
df4 = pd.DataFrame(result4)
df4.columns = ['User1', 'User2', 'Common_Fields_of_Interest', 'Common_General_Interests']
df4
```

```python
# QUERY 5: Retrieve top 3 Users who created most Posts
query5 = '''
    MATCH (p:Post)-[:CREATED_BY]->(u:User)
    WITH u, COUNT(p) AS count
    RETURN u.name, count
    ORDER BY count DESC
    LIMIT 3
'''
result5 = session.run(query5)
df5 = pd.DataFrame(result5)
df5.columns = ['User', 'Number_of_Posts']
df5
```
